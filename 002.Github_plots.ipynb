# Github Visualization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## 1st Level Users -> Repo -> Contributors -> 2nd Level Users

firstLevelUsersDf = pd.read_json("results/GithubEPFLFirstLevelUsers.json", orient='records')

firstLevelUsersRepoDf = pd.read_json("results/GithubFirstLevelUsersRepositories.json", orient='records')

firstLevelUsersRepoContributorsDf = pd.read_json("results/GithubEPFLFirstLevelUsersRepositoriesContributors.json", orient='records')

secondLevelUsersDf = pd.read_json("results/GithubEPFLSecondLevelUsers.json", orient='records')

## 1st Level Repositories -> Owners -> Contributors -> 2nd Level Repositories

firstLevelReposDf = pd.read_json("results/GithubEPFLRepositories.json", orient='records')

firstLevelReposOwnersDf =  pd.read_json("results/GithubEPFLRepoOwners.json", orient='records')

firstLevelContributorsDf = pd.read_json("results/GithubEPFLRepositoriesContributors.json", orient="records")

secondLevelRepoUsersDf = pd.read_json("results/GithubEPFLRepositoriesUniqueContributors.json", orient="records")
```

# First Level Users

```python
# Type vs Number
countDf = firstLevelUsersDf.groupby(['type']).count()

ax = sns.barplot(data=countDf, x=countDf.index, y='id')
ax.set(xlabel='Account type', ylabel='Number of Users')
plt.title("First level users after 'EPFL' search.")
plt.show()
```

## First Level Users Repositories

```python
# created_at vs updated_at
# description word cloud
# number of forks
# language diversity
# Number of watchers, stargazers_count, and forks count
# Number of contributors
# Number of repositories by type

firstLevelUsersRepoDf.keys()
```

```python
firstLevelUsersRepoDf.sort_values(by=['stargazers_count'], ascending=False).iloc[:10]
```

```python
forkDf = firstLevelUsersRepoDf.groupby('fork').count()
forkDf.index = ['False', 'True']

ax = sns.barplot(data=forkDf, x=forkDf.index, y='node_id')
ax.set(xlabel='Fork?', ylabel='Number of Repositories')
plt.title("Type of Repository")
plt.show()
```

```python
sns.histplot(data=firstLevelUsersRepoDf, x="created_at")
plt.show()

sns.histplot(data=firstLevelUsersRepoDf, x="updated_at")
plt.show()
```

```python
metricsDf = firstLevelUsersRepoDf[['watchers_count','stargazers_count','forks_count']].melt()

ax = sns.barplot(data=metricsDf, x='variable', y='value', estimator=sum, ci=None)
ax.set(xlabel='Metric', ylabel='Value')
plt.title("Metric of repositories")
plt.show()
```

```python
languageDf = firstLevelUsersRepoDf.groupby(['language']).count().sort_values(by=['id'], ascending=False).iloc[:12]

ax = sns.barplot(data=languageDf, x=languageDf.index, y='id')
ax.set(xlabel='Language type', ylabel='Number of Repositories')
plt.title("Language diversity.")
plt.xticks(rotation=90)
plt.show()
```

```python
# Number of repositories by type
repoAndOwners = pd.merge(firstLevelUsersDf, firstLevelUsersRepoDf, left_on='login', right_on='owner_login', suffixes=('_usersDB','_reposDB'))
#repoAndOwners.keys()
nReposDf = repoAndOwners.groupby(['type']).count()

ax = sns.barplot(data=nReposDf, x=nReposDf.index, y='id_usersDB')
ax.set(xlabel='Account type', ylabel='Number of Repositories')
plt.title("Number of Repositories by Type")
plt.show()
```
