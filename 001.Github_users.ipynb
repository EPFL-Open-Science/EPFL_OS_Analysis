## Extraction of users with EPFL mention

In this script we are able to extract users that are giving when search for EPFL.

```python
from dotenv import dotenv_values

config = dotenv_values(".env")
```

```python
%%time
import requests
import time
import pandas as pd

h = {'Authorization': config['GITHUB_TOKEN']}

## Include Token as a secret!

sorts = ['followers', 'repositories', 'joined', 'best-match']
orders = ['asc','desc']
items = []

for o in orders:
    for s in sorts:
        #time.sleep(60)
        cont = True
        page = 1
        while cont == True:
            print(f"Page {page}")
            #time.sleep(1.5)
            url = f"https://api.github.com/search/users?q=EPFl&page={page}&sort={s}&order={o}"
            r = requests.get(url, headers=h)

            if 'message' in r.json().keys():
                if 'API rate limit exceeded' in r.json()['message']:
                    print(r.json())
                    time.sleep(60)
                else:
                    print(r.json())
                    cont = False
            else:
                i = r.json()['items']
                items.extend(i)
                page += 1
                
itemssimple = [ {'id': i['id'], 
                 'login': i['login'], 
                 'node_id': i['node_id'],
                 'url': i['url'],
                 'html_url': i['html_url'], 
                 'followers_url': i['followers_url'],
                 'repos_url': i['repos_url'],
                 'type': i['type'],
                 'score': i['score']} for i in items]

df = pd.DataFrame(itemssimple)
df.drop_duplicates(keep='first', inplace=True)

print(f"Found {df.shape[0]} unique users") #882 #894 # 1241
```

## Query of Repositories and Contributors to first level users

```python
%%time
import requests
import time

h = {'Authorization': config['GITHUB_TOKEN']}

def getRepoAndContributors(df, verbose=False, sleep_time=60):
    repoDf = {'id': [],
            'node_id': [],
            'name': [],
            'full_name': [],
            'created_at': [],
            'updated_at': [],
            'description': [],
            'fork': [],
            'url': [],
            'html_url': [],
            'created_at': [],
            'updated_at': [],
            'git_url': [],
            'homepage': [],
            'stargazers_count':[],
            'watchers_count': [],
            'contributors_url': [],
            'language': [],
            'forks_count': [],
            'license': [],
            'topics': [],
            'forks': [],
            'open_issues': [],
            'watchers': [],
            'default_branch' :[],
            'owner_login':[], 
            'owner_url':[]}
            #'n_contributors':[]}

    conDf = {'repo_url': [], 'login': [], 'url': [], 'id': []}
    
    items = []

    failed = []
    failedContributors = []

    for i, row in df.iterrows():
        print(i[1])

        try:
            outRep = requests.get(row['repos_url'], headers=h).json()
            if type(outRep) != list:
                if outRep:
                    if 'message' in outRep.keys():
                        if 'API rate limit exceeded' in outRep['message']:
                            if verbose:
                                print(outRep)
                            time.sleep(sleep_time)
                            outRep = requests.get(row['repos_url']).json()
                        elif 'Server Error' in outRep['message']:
                            if verbose:
                                print(outRep)
                            time.sleep(sleep_time)
                            outRep = requests.get(row['repos_url']).json()
                        else:
                            if verbose:
                                print(row['repos_url'])
                                print(outRep)
                            failed.append(i)
                            continue
                else:
                    if verbose:
                        print(outRep)
                    failed.append(i)
                    continue

            for out in outRep:
                try:
                    repoDfKeys = [k for k in repoDf.keys() if k != 'owner_login' and k != 'owner_url' and k != 'n_contributors']
                    for k in repoDfKeys:
                        repoDf[k].append(out[k])

                    repoDf['owner_login'].append(out['owner']['login'])
                    repoDf['owner_url'].append(out['owner']['url'])
                    
                    if verbose:
                        print(out['contributors_url'])
                    outCon = requests.get(out['contributors_url'], headers=h)

                    if outCon.status_code == 200:
                        outCon = outCon.json()
                        if len(outCon) > 0:
#                             repoDf['n_contributors'].append(len(outCon))
                            for con in outCon:
                                conDf['repo_url'].append(out['html_url'])
                                conDf['login'].append(con['login'])
                                conDf['url'].append(con['url'])
                                conDf['id'].append(con['id'])
#                         else:
#                             repoDf['n_contributors'].append(0)
                    else:
                        if verbose:
                            print(f"Contributors SC: {outCon.status_code}")
                            print(outCon)
                            
                        failedContributors.append(i)
#                         repoDf['n_contributors'].append(None)
                        
                except Exception as e:
                    if verbose:
                        print(e)
                    failed.append(i)
                    failedContributors.append(i)
                    
        except Exception as e:
            if verbose:
                print(e)
            failed.append(i)
            failedContributors.append(i)
            
    return repoDf, conDf, failed, failedContributors
```

```python
%%time
repoDf, conDf, failed, failedContributors = getRepoAndContributors(df, verbose=True, sleep_time=2)
```

## Extraction of Creators of Repositories with EPFL mention.

In this script we are able to extract repositories that are giving when search for EPFL.

```python
%%time
import requests
import time
import pandas as pd

h = {'Authorization': config['GITHUB_TOKEN']}

sorts = ['stars', 'best-match', 'help-wanted-issues', 'forks', 'updated']
orders = ['asc','desc']
items = []

for o in orders:
    for s in sorts: 
        cont = True
        page = 1
        while cont == True:
            print(page)
            time.sleep(1)
            url = f"https://api.github.com/search/repositories?q=EPFl&page={page}&sort={s}&order={o}"
            r = requests.get(url, headers=h)

            if 'message' in r.json().keys():
                print(r.json())
                cont = False
            else:
                i = r.json()['items']
                items.extend(i)
                page += 1
                
itemssimple = [ {'id': i['id']} for i in items]

df = pd.DataFrame(itemssimple)
df.drop_duplicates(keep='first', inplace=True)

print(f"Found {df.shape[0]} unique repositories")
```

```python
repoDf = {'id': [],
        'node_id': [],
        'name': [],
        'full_name': [],
        'created_at': [],
        'updated_at': [],
        'description': [],
        'fork': [],
        'url': [],
        'html_url': [],
        'created_at': [],
        'updated_at': [],
        'git_url': [],
        'homepage': [],
        'stargazers_count':[],
        'watchers_count': [],
        'contributors_url': [],
        'language': [],
        'forks_count': [],
        'license': [],
        'topics': [],
        'forks': [],
        'open_issues': [],
        'watchers': [],
        'default_branch' :[],
        'owner_login':[], 
        'owner_url':[]}


for item in items:
    repoDfKeys = [k for k in repoDf.keys() if k != 'owner_login' and k != 'owner_url']
    for k in repoDfKeys:
        repoDf[k].append(item[k])

    repoDf['owner_login'].append(item['owner']['login'])
    repoDf['owner_url'].append(item['owner']['url'])
    
allRepositories = pd.DataFrame(repoDf).drop_duplicates(subset=['id'],keep='first').reset_index(drop=True)

licenses = []
for i, row in allRepositories['license'].items():
    if row == None:
        licenses.append(None)
    else:
        licenses.append(row['key'])
allRepositories['license'] = licenses
allRepositories
```

```python
# Export of Repositories DF
allRepositories.to_excel("results/GithubEPFLRepositories.xlsx")  
allRepositories.to_json("results/GithubEPFLRepositories.json", orient="records")
```
